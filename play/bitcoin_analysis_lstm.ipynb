```python
import okx.MarketData as MarketData, time, pandas as pd, numpy as np
from IPython.display import display
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler

flag = "0"  # Production trading:0 , demo trading:1
marketDataAPI =  MarketData.MarketAPI(flag=flag)
```

```python
""" Fetch Short Term Data """
# Retrieve the candlestick charts
# This endpoint can retrieve the latest 1,440 data entries. 
# Charts are returned in groups based on the requested bar.
instId="BTC-USDT"
bar = "1D"
limit = "300"
total_data = []
current_ts = None

while True:
    # Retrieve candlestick data
    if current_ts:
        result = marketDataAPI.get_candlesticks(instId=instId, bar=bar, limit=limit, after=current_ts)
    else:
        result = marketDataAPI.get_candlesticks(instId=instId, bar=bar, limit=limit)

    # Extract the data and add to the total_data list
    data = result['data']
    
    # Break the loop if no more data is retrieved
    if not data:
        break
    
    # Add retrieved data to total_data
    total_data.extend(data)
    
    # Update current_ts to the timestamp of the last retrieved dataset
    current_ts = data[-1][0]  # The structure of 'data' is: [[...], [...], [...], ..., [...]]
    
    # Pause to avoid hitting rate limits
    time.sleep(1) # Adjust the sleep time as needed

cols = ['Timestamp', 'Open', 'High', 'Low', 'Close', 'Vol_BTC', 'Vol_USDT', 'Vol__USDT', 'K-line_completion']
df_st = pd.DataFrame(total_data, columns=cols)

df_st['Timestamp'] = pd.to_numeric(df_st['Timestamp'])
df_st['Timestamp'] = pd.to_datetime(df_st['Timestamp'], unit='ms')

display(df_st)
df_st.to_csv('BTC_USDT-ST-1D.csv', index=False)
```

```python
""" Fetch Long Term Data """
# Retrieve history candlestick charts from recent years
# But it only retrieves last 3 months data for 1s candlestick.
instId="BTC-USDT"
bar = "1D"
limit = "100"
total_data = []
current_ts = None

while True:
    # Retrieve candlestick data
    if current_ts:
        result = marketDataAPI.get_history_candlesticks(instId=instId, bar=bar, limit=limit, after=current_ts)
    else:
        result = marketDataAPI.get_history_candlesticks(instId=instId, bar=bar, limit=limit)

    # Extract the data and add to the total_data list
    data = result['data']
    
    # Break the loop if no more data is retrieved
    if not data:
        break
    
    # Add retrieved data to total_data
    total_data.extend(data)
    
    # Update current_ts to the timestamp of the last retrieved dataset
    current_ts = data[-1][0]  # The structure of 'data' is: [[...], [...], [...], ..., [...]]
    
    # Pause to avoid hitting rate limits
    time.sleep(1) # Adjust the sleep time as needed

cols = ['Timestamp', 'Open', 'High', 'Low', 'Close', 'Vol_BTC', 'Vol_USDT', 'Vol__USDT', 'K-line_completion']
df_lt = pd.DataFrame(total_data, columns=cols)

df_lt['Timestamp'] = pd.to_numeric(df_lt['Timestamp'])
df_lt['Timestamp'] = pd.to_datetime(df_lt['Timestamp'], unit='ms')

display(df_lt)
df_lt.to_csv('BTC_USDT-LT-1D.csv', index=False)
```

## Detect outliers / sudden spikes in 'Close'

```python
def detect(signal, threshold = 2.0):
    detected = []
    for i in range(len(signal)):
        if np.abs(signal[i]) > threshold:
            detected.append(i)
    return detected

signal = np.copy(df['Close'].values)
std_signal = (signal - np.mean(signal)) / np.std(signal)
s = pd.Series(std_signal)
s.describe(percentiles=[0.25, 0.5, 0.75, 0.95])
```

```python
outliers = detect(std_signal, 1.3)

# Create the plotly figure
fig = go.Figure()
# Add the signal trace
fig.add_trace(go.Scatter(
    x=np.arange(len(signal)),
    y=signal,
    mode='lines',
    line=dict(color='grey', width=1),
    name='signal'
))
# Add the outliers trace
fig.add_trace(go.Scatter(
    x=np.arange(len(signal))[outliers],
    y=signal[outliers],
    mode='markers',
    marker=dict(color='red', symbol='x', size=2),
    name='outliers'
))
# Update x-axis ticks
fig.update_xaxes(
    tickvals=np.arange(len(signal))[::15],  # [start:stop:step]
    ticktext=df['Timestamp'][::15].astype(str),
    tickangle=-90,
    autorange="reversed"
)
# Set the figure size / style
fig.update_layout(
    width=2560,
    # height=1440,
    template='plotly_dark',
)
# Show the figure
fig.show()
```
